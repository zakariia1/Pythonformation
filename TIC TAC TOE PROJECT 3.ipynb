```python
import tkinter as tk
from tkinter import messagebox

# Initialize variables
root = tk.Tk()
root.title("TIC-TAC-TOE")

# Global variables
count = 0
mark = 'X'
digits = [1, 2, 3, 4, 5, 6, 7, 8, 9]

# Function to check for a win
def win(player_mark):
    # Check rows, columns, and diagonals for a win
    return (
        (buttons[0]['text'] == buttons[1]['text'] == buttons[2]['text'] == player_mark) or
        (buttons[3]['text'] == buttons[4]['text'] == buttons[5]['text'] == player_mark) or
        (buttons[6]['text'] == buttons[7]['text'] == buttons[8]['text'] == player_mark) or
        (buttons[0]['text'] == buttons[3]['text'] == buttons[6]['text'] == player_mark) or
        (buttons[1]['text'] == buttons[4]['text'] == buttons[7]['text'] == player_mark) or
        (buttons[2]['text'] == buttons[5]['text'] == buttons[8]['text'] == player_mark) or
        (buttons[0]['text'] == buttons[4]['text'] == buttons[8]['text'] == player_mark) or
        (buttons[2]['text'] == buttons[4]['text'] == buttons[6]['text'] == player_mark)
    )

# Function to handle button clicks
def button_click(idx):
    global count, mark, digits
    if digits[idx] and buttons[idx]['text'] == "":
        digits[idx] = 0
        buttons[idx]['text'] = mark
        count += 1
        if count >= 5:
            if win(mark):
                messagebox.showinfo("TIC-TAC-TOE", mark + " wins!")
                root.destroy()
            elif count == 9:
                messagebox.showinfo("TIC-TAC-TOE", "Match Tied")
                root.destroy()
        mark = 'X' if mark == 'O' else 'O'

# Create labels for player symbols
label_x = tk.Label(root, text="Player X", font=("Helvetica", 16))
label_x.grid(row=0, column=0)
label_o = tk.Label(root, text="Player O", font=("Helvetica", 16))
label_o.grid(row=0, column=1)

# Create buttons for the cells
buttons = [tk.Button(root, text="", font=("Helvetica", 16), width=8, height=3, command=lambda i=i: button_click(i)) for i in range(9)]

# Position the buttons using the grid method
for i, button in enumerate(buttons):
    button.grid(row=i // 3 + 1, column=i % 3)

# Run the Tkinter main loop
root.mainloop()
```
